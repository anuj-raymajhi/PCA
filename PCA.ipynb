```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#that samples 20 points from a normal distribution

samp1 = np.random.standard_normal((20,2))
samp2 = np.random.standard_normal((20,2))
```

```python
print(samp1)
```

```python
plt.scatter(samp1[:,0],samp1[:,1], linewidths = 0.1)
```

```python

samp1.mean()
```

```python
samp1.std()
```

```python
mat1 = np.random.rand(2,2)
```

```python
mul=np.matmul(samp1,mat1)
```

```python
plt.scatter(mul[:,0],mul[:,1],linewidths=0.05)
```

```python
print(np.var(mul[:,0]))
print(np.var(mul[:,1]))
```

```python
#variance along the dominant direction and direction perpendicular to it

var_mat = np.cov(mul.T)
print(var_mat)
```

```python
#calculating eigen values
eigval, eigvec= np.linalg.eig(var_mat)

eigval, eigvec
```

```python
#calculating proportion of variance
pov1 = eigval[0] / np.sum(eigval)
pov2 = eigval[1] / np.sum(eigval)

pov1, pov2
```

```python
#Y = PX
Y = np.matmul(eigvec.T, mul.T).T
Y
```

```python
plt.scatter(Y[:,0],Y[:,1])
plt.grid()
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.axhline(y = 0, color = 'r')
plt.axvline(x=0, color = 'r')
```

```python
Sy = np.cov(Y.T)
Sy
```

```python
# 1D plot using only the largest
print(np.var(Y[:,0]))
print(np.var(Y[:,1]))
```
