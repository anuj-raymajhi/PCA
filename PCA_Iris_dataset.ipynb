```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df['target'] = pd.Series(iris.target)
```

```python
X = iris_df.iloc[:,0:4].values
print(np.mean(X), np.std(X))
```

```python
X_strd = StandardScaler().fit_transform(X)
print(np.mean(X_strd), np.std(X_strd))
```

```python
varMat = np.cov(X_strd.T)
varMat
```

```python
eigVal, eigVec = np.linalg.eig(varMat)
eigVal, eigVec

# The normalized (unit “length”) eigenvectors, such that the column eigenvectors[:,i] 
# is the eigenvector corresponding to the eigenvalue eigenvalues[i]
```

```python
pov1 = eigVal[0]/np.sum(eigVal)
pov2 = eigVal[1]/np.sum(eigVal)
pov3 = eigVal[2]/np.sum(eigVal)
pov4 = eigVal[3]/np.sum(eigVal)

pov1, pov2, pov3, pov4
```

```python
eigPairs = [(np.abs(eigVal[i]), eigVec[:,i]) for i in range(len(eigVal))]
eigPairs
```

## PCA Reduction to two dimension

### Preserving (72.96 + 22.85) % variance

```python
matP = eigVec[:,0:2]
matP
```

```python
Y = X_strd.dot(matP)
```

```python
plt.figure()
targetNames = iris.target_names
y = iris.target
for c, i, targetName in zip("rgb", [0,1,2], targetNames):
    plt.scatter(Y[y==i, 0], Y[y==i, 1], c=c, label=targetName)
plt.xlabel("Principle Component 1")
plt.ylabel("Principle Component 2")
plt.legend()
plt.title('PCA of Iris Dataset')
plt.show()
```

```python
matP1 = np.column_stack((eigVec[:,0],eigVec[:,2]))
matP1
```

## PCA Reduction to 3 Dimensions

```python
from ipywidgets import interact, FloatSlider
```

```python
matQ = np.column_stack((eigVec[:,0], eigVec[:,1], eigVec[:,2]))
matQ
```

```python
Z = X_strd.dot(matQ)
Z[0:5]
```

```python
def threeD(e,z):
    fig = plt.figure(1, figsize=(8,6))
    ax = fig.add_subplot(111, projection="3d", elev = e, azim = z)
    for c, i, targetName in zip("rgb", [0,1,2], targetNames):
        ax.scatter(Y1[y==i, 0], Y1[y==i, 1], Z[y==i, 2], c=c, label=targetName)
    ax.set_xlabel("Principle Component 1")
    ax.xaxis.set_ticklabels([])
    ax.set_ylabel("Principle Component 2")
    ax.yaxis.set_ticklabels([])
    ax.set_zlabel("Principle Component 3")
    ax.zaxis.set_ticklabels([])
    ax.legend()
    ax.set_title('PCA of Iris Dataset')
```

```python
interact(threeD, e = FloatSlider(min=0, max=360, step = 0.01), z = FloatSlider(min=0, max=360, step = 0.01))
```
